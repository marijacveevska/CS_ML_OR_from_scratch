# Chapter 3. Dictionaries and Sets

Dicts and Sets share the same underlying engine — a <b>hash table</b>\
which is why both give you average O(1) lookup, insertion, and membership testing, and why both require their elements/keys to be hashable.

## Hash tables and Hash functions

- <b>Hash function</b>: The formula/calculator that takes any piece of data — a word, a number, a whole sentence, anything — and turns it into a fixed-size number.
- <b>Hash number</b>: the output — the number that comes out the other end when you run something through the hash function.
- <b>Hashing</b>: the act of running something through the formula to get its number. "Hashing the word 'apple'" means "computing its hash number."
- <b>Hash table / Hash map</b>: the actual storage structure. Picture a long row of empty numbered slots (boxes), like 1,000 numbered mailboxes in a row.
To store something:
    - Run it through the formula → get a number.
    - Use that number to pick which mailbox to put it in (e.g., number mod 1000, to fit however many mailboxes you have).
    - Drop it in that mailbox.

In Python: <b>dict and set are hash tables/hash maps</b>. \
hash(x) is the function that does the hashing. Whatever number comes out is the hash number. And a str, int, or tuple is hashable (usable as a dict key or set member) because none of them can be modified after creation.\
While a list is not hashable, because you can modify it, which would break the mailbox system.


## Dictionaries
<b>"What is a dict, conceptually?" </b>→ It's a mapping — an object that connects keys to values. This describes what it does (its interface/behavior).\
<b>"How is a dict built internally, to make that fast?"</b> → It's implemented as a hash table (a.k.a. hash map — these two are just synonyms for the same underlying data structure, no difference between them). This describes how it's built (its implementation).

## Dict Comprehensions
build a dict directly from any iterable of key-value pairs.\
{key: value for item in iterable.items}


```python
dial_codes = [(880,'Bangladesh'),(55,'Brazil'),(86,'China'),(91,'India'),(62,'Indonesia'),
(81,'Japan'),(234,'Nigeria'),(92,'Pakistan'),(7,'Russia'),(1,'US')]
country_dial = {country:code for code,country in dial_codes}

print(country_dial)
country_dial = {code:country.upper() for country,code in sorted(country_dial.items())}
print(country_dial)

print(sorted(country_dial.items()))
```

### Unpacking Mappings with **

In Ch 2 we've already used **kwargs to gather keyword arguments into a dict inside a function.\
This section shows the reverse direction — using ** to spread a dict's contents back out, in two contexts.


```python
def dump(**kwargs):
    return kwargs

dict1 = {'x':1}
y=2
dict2 = {'z':3}

dump(**dict1,y=y,**dict2)
```

### Merging Mappings with   |

The values with the same keys will get overriden.The later occurrences owerwrite the previous ones.

```python
dict1 = {'a': 1, 'b': 3}
dict2 = {'c': 5, 'd': 7, 'e': 9}

dict3 = dict1|dict2

print(dict3)
```

## Set Theory

A set is a collection of unique, unordered, hashable objects. A basic use case is removing duplication:

```python
l = ['spam', 'spam', 'eggs', 'spam', 'bacon', 'eggs']
print(set(l))
list(set(l))
```

### Set comprehention

The distinguishing factor between set and dict comprehention is whether there is : inside the brackets\
<b>Set</b>\
{expr for item in iterable}\
<b>Dict</b>\
{word: len(word) for word in words}

## Set Operations

a | b ; .union() ; everything in either\
a & b ; .intersection() ; only what's in both\
a - b ; .difference() ; in a but not b\
a ^ b ; .symmetric_difference() ; in one or the other, but not both
