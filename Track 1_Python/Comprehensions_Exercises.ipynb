# Comprehensions deep dive

1. Flatten a list of lists (e.g. `[[1,2],[3,4]]` → `[1,2,3,4]`).

```python
nested = [[1, 2], [3, 4], [5, 6]]
flatten = [item for nest in nested for item in nest]

print(flatten)
```

2. Build a dict `{n: n**2}` for even numbers only, from `range(20)`.

```python
dict1 = {n:n**2 for n in range(20) if n%2==0}
print(dict1)
```

3. Transpose a matrix (list of lists) using a **nested** comprehension.

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

column = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(column)

```

4. Given a string, build `{char: count}` frequency dict.

```python
text = "mississippi"
print(text.count("m"))

dict_char = {let:text.count(let) for let in text}
print(dict_char)
```

5. Cartesian product: given `colors = ['red','blue']` and `sizes = ['S','M']`, produce all `(color, size)` tuples.

```python
colors = ['red', 'blue']
sizes = ['S', 'M']

cartesian_list = [(col,size) for col in colors for size in sizes]
print(cartesian_list)
```

6. Given a list of dicts (e.g. `[{'name':'a','age':30}, ...]`), extract a list of just the names where age > 25.

```python
people = [
    {'name': 'alice', 'age': 30},
    {'name': 'bob', 'age': 22},
    {'name': 'carol', 'age': 41},
    {'name': 'dave', 'age': 19},
]

list_name = [dc["name"] for dc in people if dc["age"]>25]
print(list_name)
```
