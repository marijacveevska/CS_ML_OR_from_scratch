# Decorators and Closures

<b>First-class functions</b> (the underlying property of Python)\
        ↓ makes possible\
<b>Closures</b> (inner function + remembers outer variable, once outer scope is gone)\
        ↓ specific use case of\
<b>Decorators</b> (a closure that wraps a function, applied via @syntax)

# Closures
<b>Closures</b> — a function that remembers where it came from.\
A closure is what you get when an inner function uses a variable from its enclosing function, and that inner function gets returned/kept around afterward.

```python
def make_multiplier(n):
    def multiply(x):
        return x * n      # <- uses 'n' from the OUTER function
    return multiply         # <- returns the inner function

triple = make_multiplier(3)
triple(10)   # 30
```

# Decorators
<b>Decorators</b> — a specific job that closures are used for.\
A decorator is a closure, used for one particular purpose: wrapping an existing function to add behavior around it, without changing its code.\
A decorator is a callable that takes another function as argument (the decorated function).\
A decorator may perform some processing with the decorated function, and returns it or replaces it with another function or callable object.

Why use decorators:
- separation of ideas
- reusability

```python
def shout(func):              # takes a function IN
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result.upper()
    return wrapper              # returns a function OUT

@shout
def greet():
    return "hi"

greet()   # 'HI'
```

### Problem without a Decorator

```python
def run_battery_optimization(demand_profile):
    # pretend this does real work
    total = sum(demand_profile) * 1000
    return total
```

```python
import time

start = time.time()
result = run_battery_optimization([10, 20, 30])
end = time.time()
print(f"Took {end - start:.4f} seconds")
print(result)
```

### Problem with a Decorator

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)       # actually run the real function
        end = time.time()
        print(f"{func.__name__} took {end - start:.4f} seconds")
        return result                          # don't forget to return the real result!
    return wrapper

@timer
def run_battery_optimization(demand_profile):
    total = sum(demand_profile) * 1000
    return total
```

```python
run_battery_optimization([10, 20, 30])
# "run_battery_optimization took 0.0001 seconds"
# 60000   <- still returns the actual answer too
```

# EXERCISES

1. `@timer` — prints how long the wrapped function took to run.

```python
import time

def timer(func):
    def wrapper(*args,**kwargs):
        start = time.time()
        result = func(*args,**kwargs)
        end = time.time()
        print(f"{func.__name__} took {end - start:.4f} seconds")
        return result                    
    return wrapper


@timer
def slow_add(a, b):
    time.sleep(1)   # simulate work
    return a + b

slow_add(2, 3)
```

2. `@log_calls` — prints the function name and arguments every time it's called.

```python
def log_calls(func):
    def wrapper(*args,**kwargs):
        print(f"function name: {func.__name__}, arguments: {args},{kwargs}")
        result = func(*args,**kwargs)
        return result
    return wrapper

@log_calls
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

greet("Marija", greeting="Hi")
```

3. `@memoize` — caches results by argument, so a slow recursive function (test with a naive `fibonacci(n)`) speeds up dramatically on repeat calls.

```python
def memoize(func):
    cache = {}
    def wrapper(*args,**kwargs):
        if args in cache:
            return cache[args]
        else:
            result = func(*args, **kwargs)
            cache[args] = result
            return result

    return wrapper



@memoize
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

start = time.time()
result = fibonacci(30)
end = time.time()
print(f"Took {end - start:.4f} seconds")
print(result)
```
