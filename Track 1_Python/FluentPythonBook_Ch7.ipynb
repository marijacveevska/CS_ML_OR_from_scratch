# Functions as First-class Object

### What "object" means in programming
An object is a chunk of data in memory that has:

- An identity — it exists somewhere specific in memory (you've already seen this — id()).
- A type — it knows what kind of thing it is (int, str, list, function, ...).
- Some content/behavior — data it holds, and things you can do to it (methods, operations).

```python
x = 5
print(type(x))     # <class 'int'>
print(id(x))       # some memory address
print(x.bit_length())   # 3 — ints even have methods you can call!
```

```python
print('hello'.upper())     # you're calling a method ON the string object

list1 = [1,2,3]
list1.append(4)

print(list1)   # you're calling a method ON the list object
```

### Functions are also objects

```python
def greet():
    """Say hello."""
    return "hi"

print(greet())

print(type(greet))        # <class 'function'>
print(id(greet))          # some memory address — it lives in memory, just like [1,2,3] does
print(greet.__doc__)      # 'Say hello.'  — it even has attached data (its docstring)
```

Because functions are objects, they can do everything other objects can do:

```python
my_func = greet        # assign it to a variable (no parentheses — don't call it, just reference it)
print(my_func)          # <function greet at 0x...> — you're looking at the object itself

functions_list = [greet, print, len]   # store functions in a list, just like numbers or strings
print(functions_list)

def call_it(f):          # pass a function in as an argument, like any other value
    return f()

call_it(greet)             # 'hi'
```

## Higher Order Functions

A function that takes a function as argument or returns a function as the result is a higher-order function.\
example: sorted()

```python
fruits = ['strawberry', 'fig', 'apple', 'cherry', 'raspberry','banana']
sorted(fruits,key=len)
```

```python
def reverse(word):
    return word[::-1]

print(reverse("stockholm"))

sorted(fruits,key=reverse)
```

```python
# Lambda function

fruits = ['strawberry', 'fig', 'apple', 'cherry', 'raspberry',
'banana']
sorted(fruits, key=lambda word: word[::-1])
```

# Callable objects

It's a simple yes/no test: "can I put () after this object and have it run?"\
This chapter is listing every category of Python object that answers True to that question.

```python
print(callable(len))     # True  — len is a function, can be called with ()
print(callable(5))       # False — you can't do 5() 
callable('hello')        # False — you can't do 'hello'()
```

| Callable type | How it's created | Example |
|---|---|---|
| User-defined function | `def` or `lambda` | `def greet(): ...` |
| Built-in function | Implemented in C, ships with Python | `len()`, `print()` |
| Built-in method | Implemented in C, attached to a built-in type | `dict.get`, `list.append` |
| Method | Function defined inside a class body | `self.some_method()` |
| Class | `class` statement; calling it creates an instance | `list()`, `MyClass()` |
| Class instance (with `__call__`) | Instance of a class that defines `__call__` | `my_obj()` |
| Generator function | Function/method containing `yield` | Returns a generator object when called |
| Native coroutine function | `async def` (no `yield`) | Returns a coroutine object when called |
| Async generator function | `async def` containing `yield` | Returns an async generator, used with `async for` |

## Freezing Arguments with functools.partial

```python
from operator import mul

mul(3, 7)   # 21

# mul is just a function version of the * operator — mul(a, b) is the same as a * b.

from functools import partial

triple = partial(mul, 3)

# Read this as: "give me a new function, based on mul, where the first argument is already permanently set to 3." 
# triple is now effectively def triple(y): return 3 * y — except you didn't write that function by hand; partial built it for you.

triple(5) # (3 * 5) = 15
```
